```python
# формируем перечень всех требуемых показателей ('BILL_*', 'PAY_*') + "default payment next month"
CN = df.columns
CN = CN[(CN.str.contains('BILL_')) | (CN.str.contains('PAY_')) | (CN.str.contains('default.payment.next.month'))] 
```

```python
#группируем и выводим медиану по требуемым параметрам
df[CN].groupby('default.payment.next.month').median()
```
